# Scraping URLs 

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse

def get_links_from_page(url, base_domain, keywords):
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        links = soup.find_all('a', href=True)
        absolute_links = [urljoin(url, link['href']) for link in links 
                          if urlparse(urljoin(url, link['href'])).netloc == base_domain
                          and any(segment in urlparse(urljoin(url, link['href'])).path.split('/') 
                                  for segment in keywords)]
        return absolute_links
    except Exception as e:
        print(f"Error fetching links from {url}: {e}")
        return []

def get_all_links(base_url, segments):  
    base_domain = urlparse(base_url).netloc
    all_links = set()
    links_to_visit = set([base_url])
    

    while links_to_visit:  
        current_url = links_to_visit.pop()
        #print(f"Visiting {current_url}")
        links_on_page = get_links_from_page(current_url, base_domain, segments)
        new_links = set(links_on_page) - all_links
        all_links.update(links_on_page)
        links_to_visit.update(new_links)

    return all_links

if __name__ == "__main__":
    base_url = input("Enter the base URL to crawl: ")
    keywords = ['www.mckinsey.com',
 'www.ey.com',
 'kpmg.com',
 'www.pwc.com.au',
 'www.deloitte.com',
 'media',
 'news',
 'agenda',
 'news-centre',
 'media-releases',
 'www.ergonomics.org.au',
 'acif-news',
 'www.constructors.com.au',
 'Media_Releases',
 'media-release',
 'healthcareworkersaustralia.com',
 'masterbuilders.com.au',
 'work-health-and-safety',
 'theconversation.com',
 'www.cdc.gov',
 'news-and-events',
 'news-and-events/safety-alerts',
 'news-and-alerts',
 'news_releases',
 'www.mom.gov.sg',
 'announcements',
 'publications',
 'news-and-media/safety-alerts',
 'news-and-media',
 'safetyatworkblog.com',
 'news-item',
 'All-News',
 'blog',
 'whsmatters.com.au',
 'www.actionohs.com.au',
 'press',
 'www.amwu.org.au',
 'www.nswnma.asn.au',
 'stay-informed',
 'www.ioshmagazine.com',
 'news-and-opinion']
    all_links = get_all_links(base_url, keywords)
    print("\nAll Links:")
    for link in all_links:
        print(link)
```

# Newspaper

```python
import pandas as pd
from newspaper import Article, Config
import nltk
import dateparser

# Download NLTK punkt tokenizer
nltk.download('punkt')

result = []
user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:78.0) Gecko/20100101 Firefox/78.0'
config = Config()
config.browser_user_agent = user_agent
config.request_timeout = 10

for link in all_links:
    try:
        article = Article(link, config=config)
        article.download()
        article.parse()
        article.nlp()

        headline = article.title
        content = article.text
        date = article.publish_date
        url = article.url
        result.append([headline, content, date, url])
    except Exception as e:
        pass  # Ignore the error silently

df = pd.DataFrame(result, columns=['Headline', 'Content', 'Date', 'Reference URL'])
unique_df = df.drop_duplicates(subset=['Headline', 'Date'])

unique_df
```

# Date Range filter

```python
import pandas as pd

# Assuming unique_df is your DataFrame containing the article data
# Convert the 'Date' column to datetime if it's not already in datetime format
unique_df['Date'] = pd.to_datetime(unique_df['Date'], utc=True).dt.strftime('%Y-%m-%d')

# Filter out rows with null values in the 'Date' column
unique_df = unique_df.dropna(subset=['Date'])

# Prompt the user to enter the date range
start_date_input = input("Please enter the start date (YYYY-MM-DD): ")
end_date_input = input("Please enter the end date (YYYY-MM-DD): ")


try:
    # Convert user inputs to datetime objects
    start_date = pd.to_datetime(start_date_input, utc=True).strftime('%Y-%m-%d')
    end_date = pd.to_datetime(end_date_input, utc=True).strftime('%Y-%m-%d')
except ValueError:
    print("Invalid date format. Please enter dates in YYYY-MM-DD format.")
    exit()

# Filter the DataFrame based on the custom date range
filtered_df = unique_df[(unique_df['Date'] >= start_date) & (unique_df['Date'] <= end_date)]

# Display the filtered DataFrame
filtered_df

```

# Cosine Similarity

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

df = filtered_df

# Given phrase
phrase = input("Enter a keyword: ")

# Create TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit the vectorizer and transform the content column
tfidf_matrix = tfidf_vectorizer.fit_transform(df['Content'])

# Transform the phrase into a TF-IDF vector
phrase_vector = tfidf_vectorizer.transform([phrase])

# Calculate cosine similarity between each row and the phrase
similarity_scores = cosine_similarity(phrase_vector, tfidf_matrix)

# Add similarity scores to DataFrame
df['Similarity Score'] = similarity_scores.flatten()

# Convert dictionary output to DataFrame
output_df = pd.DataFrame(df)

output_df
```

```python
final_df = output_df[output_df['Similarity Score'] > 0.266]

final_df
```

```python
import google.generativeai as genai
from google.generativeai.types import HarmCategory, HarmBlockThreshold
import os
```

```python
genai.configure(api_key='get_Api_key') #refer documentaion to get the api keu for gemini pro.
```

```python
model = genai.GenerativeModel('gemini-pro')
```

```python
def generate_insights(df, text_column, insight_column):
    # Iterate over each row in the dataframe
    for index, row in df.iterrows():
        # Get text from the specified column
        text = row[text_column]
        
        # Generate insight from the text
        insight = model.generate_content("generate insight in billet points (*) with correct statstical details in maximum 100 words from text:" + text,
        safety_settings={
        HarmCategory.HARM_CATEGORY_HATE_SPEECH: HarmBlockThreshold.BLOCK_NONE,
        HarmCategory.HARM_CATEGORY_HARASSMENT: HarmBlockThreshold.BLOCK_NONE,
        HarmCategory.HARM_CATEGORY_SEXUALLY_EXPLICIT: HarmBlockThreshold.BLOCK_NONE,
        HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT: HarmBlockThreshold.BLOCK_NONE,
    }
    )
        
        # Save the insight into the new column
        df.at[index, insight_column] = insight.text
        
    return df

# Generate insights for the 'text_column' and save them into a new column named 'insight_column'
gemini = generate_insights(final_df, 'Content', 'insight_Gemini')

# Print the updated DataFrame
gemini
```

### LLaMA chat

```python
from langchain_community.chat_models import ChatOllama
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate

# Initialize Ollama
llm = ChatOllama(model="llama2")
prompt = ChatPromptTemplate.from_template("generate insight in billet points (*) with correct statstical details in maximum 100 words from from {topic}")
chain = prompt | llm | StrOutputParser()

def generate_insights(df, text_column, insight_column):
    # Iterate over each row in the dataframe
    for index, row in df.iterrows():
        # Get text from the specified column
        text = row[text_column]
        
        # Generate insight from the text
        insight = chain.invoke({"topic": text})
        
        # Save the insight into the new column
        df.at[index, insight_column] = insight
        
    return df

# Generate insights for the 'text_column' and save them into a new column named 'insight_column'
llamachat = generate_insights(final_df, 'Content', 'insight_llamachat')

# Print the updated DataFrame
llamachat
```
