# the list of href links

```python
import requests
from bs4 import BeautifulSoup

# URL of the search page
url = "https://www.ey.com/en_au/search?q=work+health+and+safety"

# Send a GET request
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find all <h3> tags
headings = soup.find_all('h3')

# Extract href links
href_links = []

for heading in headings:
    link = heading.find('a', class_='hyperlink-text-link')
    if link:
        href_links.append(link['href'])

# Print the list of href links
href_links
```

# Loop through pages to get all links

```python
import requests
from bs4 import BeautifulSoup
import re

# Base URL of the search page
base_url = "https://www.ey.com/en_au/search?q=work+health+and+safety"

# Send a GET request to get the first page
response = requests.get(base_url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find the total number of pages
pagination = soup.find('div', class_='pagination-list')
pages = pagination.find_all('a', class_='pagination-item')
total_pages = int(pages[-1].text.strip())

# List to store all href links
all_href_links = []

# Loop through each page
for page_num in range(1, total_pages + 1):
    # Construct URL for the current page
    url = f"{base_url}&p={page_num}"
    
    # Send a GET request
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find all <h3> tags
    headings = soup.find_all('h3')

    # Extract href links
    for heading in headings:
        link = heading.find('a', class_='hyperlink-text-link')
        if link:
            all_href_links.append(link['href'])

# Print the list of href links
all_href_links
```

# Code to get all the text from each website

```python
import requests
from bs4 import BeautifulSoup

# Define the URL of the website you want to scrape
url = 'https://www.ey.com/en_au/environment-health-safety/services'

# Create a session object
session = requests.Session()

# Send a GET request to the website
response = session.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the webpage
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extract all text content from the webpage
    text_content = soup.get_text()
    
    # Print out the text content
    print(text_content)
else:
    print("Failed to retrieve the webpage.") 
```

# Data cleaning 

### Remove pancuation marks

```python
import string

def remove_punctuation(text):
    # Get all punctuation marks
    punctuations = string.punctuation
    
    # Remove punctuation marks from the text
    clean_text = ''.join(char for char in text if char not in punctuations)
    
    return clean_text


cleaned_text = remove_punctuation(text_content)
print(cleaned_text)
```

### remove empty spaces

```python
import string

def remove_punctuation_and_spaces(text):
    # Get all punctuation marks
    punctuations = string.punctuation
    
    # Remove punctuation marks and empty spaces between lines
    clean_lines = []
    for line in text.split('\n'):
        clean_line = ''.join(char for char in line.strip() if char not in punctuations)
        if clean_line:  # Check if the line is not empty after removing punctuation
            clean_lines.append(clean_line)
    
    return ' '.join(clean_lines)



cleaned_text = remove_punctuation_and_spaces(text_content)
print(cleaned_text)
```

### Remove irrerelvent data from top

```python
def delete_text_until(text, marker):
    # Find the index of the marker text
    marker_index = text.find(marker)
    
    # If the marker is found, delete text before it, otherwise return original text
    if marker_index != -1:
        cleaned_text = text[marker_index + len(marker):]
    else:
        cleaned_text = text
    return cleaned_text

# Example usage:
marker = "Related topics"
cleaned_text = delete_text_until(cleaned_text, marker)
print(cleaned_text)
```

### Remove Irrelevent data from bottom

```python
def delete_text_until(text, marker):
    # Find the index of the marker text
    marker_index = text.find(marker)
    
    # If the marker is found, delete text after it, otherwise return original text
    if marker_index != -1:
        cleaned_text = text[:marker_index]
    else:
        cleaned_text = text
    return cleaned_text

marker = "Connect with us"
cleaned_text = delete_text_until(cleaned_text, marker)
print(cleaned_text)
```
